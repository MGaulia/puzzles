```python
def sortbyref(values, references):
    return [x for _, x in sorted(zip(references, values))]

def permutation(lst):
    if len(lst) == 0:
        return []
    if len(lst) == 1:
        return [lst]

    l = [] 

    for i in range(len(lst)):
        m = lst[i]
        remainderLst = lst[:i] + lst[i+1:]
        
        for p in permutation(remainderLst):
            l.append([m] + p)
    return l


def get_fun_dict(fun):
    df = {}
    for i in range(1, len(fun) + 1):
        df[i] = fun[i-1]
    return df

def get_dict_pointers(pointers):
    dict_pointers = {}
    for i in range(1, len(pointers) + 1):
        dict_pointers[i] = pointers[i-1]
    
    return dict_pointers

def get_path(start, dp):
    res = [start]
    while start != 0:
        start = dp[start]
        if start != 0:
            res.append(start)
    
    return res

def getallpaths(start_comb, dp):
    paths = []
    for i in start_comb:
        paths.append(get_path(i, dp))
    
    return paths

def getstarters(pointers):
    starters = list(range(1, len(pointers) + 1))
    for i in pointers:
        if i in starters:
            starters.remove(i)
    return starters

def get_combination_score(paths, df):
    df = get_fun_dict(fun)
    score = []
    for path in paths:
        funlist = []
        for visit in path:
            funlist.append(df[visit])
            df[visit] = 0
        score.append(max(funlist))

    return sum(score)

def get_sc(x):
    x = permutation(x)
    x = [sum(i, []) for i in x]
    return x

def solve(fun, pointers, case):
    dp = get_dict_pointers(pointers)
    df = get_fun_dict(fun)
    starters = getstarters(pointers)
    
    neibs = {}
    for visitor in starters:
        
        home = dp[visitor]
        if  home not in neibs:
            neibs[dp[visitor]] = [visitor]
        else:
            temp = neibs[home]
            temp.append(visitor)
            neibs[home] = temp
    
    for home, visitors in neibs.items():
        neibs[home] = sortbyref(visitors, [df[i] for i in visitors])
    
    result = 0    
    start_combinations = get_sc(list(neibs.values()))

    for i in start_combinations:
        allpaths = getallpaths(i, dp)

        currscore = get_combination_score(allpaths, fun)
        if currscore > result:
            result = currscore
    
    print("Case #" + str(case) + ": " + str(result))
```
